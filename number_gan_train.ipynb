```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets("../03-Convolutional-Neural-Networks/MNIST_data/",one_hot=True)
```

```python
def generator(z,reuse=None):
    with tf.variable_scope('gen',reuse=reuse):
        hidden1 = tf.layers.dense(inputs=z,units=128)
        
        alpha = 0.01
        hidden1 = tf.maximum(alpha*hidden1,hidden1)
        
        hidden2 = tf.layers.dense(inputs=hidden1,units=128)
        hidden2 = tf.maximum(alpha*hidden2,hidden2)
        
        output = tf.layers.dense(hidden2,units=28**2,activation=tf.nn.tanh)
        
        return output
```

```python
def discriminator(X,reuse=None):
    with tf.variable_scope('dis',reuse=reuse):
        hidden1 = tf.layers.dense(inputs=X,units=128)
        
        alpha = 0.01
        hidden1 = tf.maximum(alpha*hidden1,hidden1)
        
        hidden2 = tf.layers.dense(inputs=hidden1,units=128)
        hidden2 = tf.maximum(alpha*hidden2,hidden2)
        
        logits = tf.layers.dense(hidden2,units=1)
        output = tf.sigmoid(logits)
        
        return output, logits
```

```python
real_images = tf.placeholder(tf.float64,shape=[None,28**2])
z = tf.placeholder(tf.float64,shape=[None,100])
```

```python
G = generator(z)
D_output_real, D_logits_real = discriminator(real_images)
```

```python
D_out_fake, D_logits_fake = discriminator(G,reuse=True)
```

```python
def loss_func(logits,labels_in):
    return tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=logits, labels=labels_in))
```

```python
D_real_loss = loss_func(D_logits_real,tf.ones_like(D_logits_real)*0.9)
D_fake_loss = loss_func(D_logits_fake,tf.zeros_like(D_logits_fake))
```

```python
D_loss = D_real_loss + D_fake_loss
```

```python
G_loss = loss_func(D_logits_fake,tf.ones_like(D_logits_fake))
```

```python
learning_rate = 0.001
```

```python
tvars = tf.trainable_variables()

d_vars = [var for var in tvars if 'dis' in var.name]
g_vars = [var for var in tvars if 'gen' in var.name]
```

```python
D_trainer = tf.train.AdamOptimizer(learning_rate).minimize(D_loss,var_list=d_vars)
```

```python
G_trainer = tf.train.AdamOptimizer(learning_rate).minimize(G_loss,var_list=g_vars)
```

```python
batch_size = 100
epochs = 200
```

```python
init = tf.global_variables_initializer()
samples = []
saver = tf.train.Saver(var_list=g_vars)
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(epochs):
        num_batches = mnist.train.num_examples // batch_size
        
        for i in range(num_batches):
            batch = mnist.train.next_batch(batch_size)
            
            batch_images = batch[0].reshape((batch_size),28**2)
            batch_images = batch_images * 2 - 1 #Normalizing for tanh
            
            batch_z = np.random.uniform(-1,1,size=(batch_size,100))
            
            _ = sess.run(D_trainer,feed_dict={real_images:batch_images,z:batch_z})
            _ = sess.run(G_trainer,feed_dict={z:batch_z})
        
        print(f'Finished training of epoch {epoch+1}/{epochs}')
        
        sample_z = np.random.uniform(-1,1,size=(1,100))
        gen_sample = sess.run(generator(z,reuse=True),feed_dict={z:sample_z})
        
        fig, axes = plt.subplots(1, 2, figsize=(10, 10))
        
        axes[0].imshow(gen_sample.reshape(28,28),cmap='Greys')
        axes[0].set_title('Generated picture')
        
        axes[1].imshow(mnist.train.images[np.random.randint(0,mnist.train.num_examples)].reshape(28,28),cmap='Greys')
        axes[1].set_title('Random real picture')
        plt.show()
        
        samples.append(gen_sample)
        
    file_name = f'./model_{epochs}_epochs.ckpt'
    saver.save(sess, file_name)
    print(f'Saved the current training-progress to file {file_name}')
```

```python
num_samples = 10
```

```python
new_samples = []

with tf.Session() as sess:
    print('Restoring saved model.')
    saver.restore(sess,f'./model_{epochs}_epochs.ckpt')
    print(f'Generating {num_samples} new numbers.')
    for x in range(num_samples):
        sample_z = np.random.uniform(-1,1,size=(1,100))
        gen_sample = sess.run(generator(z,reuse=True),feed_dict={z:sample_z})
        new_samples.append(gen_sample)
        
    for gen_sample in new_samples:
        plt.imshow(gen_sample.reshape(28,28),cmap='Greys')
        plt.show()
```
